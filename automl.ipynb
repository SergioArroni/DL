```python
import pandas as pd
import numpy as np
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
import autokeras as ak
```

```python
seed = 0
```

```python
DATASETX = "./data/prep/HotelReservationsPreparedCleanX.csv"
DATASETY = "./data/prep/HotelReservationsY.csv"
```

```python
df_x = pd.read_csv(DATASETX)
df_y = pd.read_csv(DATASETY)
```

```python
df_x.head()
```

```python
df_y.head()
```

```python
df = pd.concat([df_x, df_y], axis=1)
df.head()
```

```python
df = shuffle(df, random_state=seed)
df.head()
```

```python
TRAIN_SIZE = 0.8
VALIDATION_SIZE = 0.1
TEST_SIZE = 0.1
```

```python
train_set, temp_set = train_test_split(df, train_size=TRAIN_SIZE, random_state=seed)

validation_set, test_set = train_test_split(temp_set, train_size=VALIDATION_SIZE / (VALIDATION_SIZE + TEST_SIZE), random_state=seed)
```

```python
X_train = train_set.drop('booking_status', axis=1)
y_train = train_set['booking_status']

X_val = validation_set.drop('booking_status', axis=1)
y_val = validation_set['booking_status']

X_test = test_set.drop('booking_status', axis=1)
y_test = test_set['booking_status']
```

```python
train_set.head()
```

```python
validation_set.head()
```

```python
test_set.head()
```

```python
print("Train: ", X_train.shape)
print("Test: ", X_test.shape)
print("Validation: ", X_val.shape)
```

```python
clf = ak.StructuredDataClassifier(seed=seed)  # Puedes ajustar max_trials según tus necesidades.
```

```python
np.object = object
```

```python
clf.fit(
    X_train,
    y_train,
    validation_data=(X_val, y_val),
    epochs=20,  # Número de épocas para entrenamiento.
)
```

```python
results = clf.evaluate(X_val, y_val)
print(results)
```

```python
best_model = clf.export_model()
best_model.summary()
```

```python
best_model.save("best_model_autokeras", save_format="tf")
```
